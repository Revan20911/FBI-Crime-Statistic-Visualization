```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)  
pd.set_option('display.expand_frame_repr', False)
pd.set_option('max_colwidth', -1)
##Importing Pandas and numpy libraries to help analyze data
# converting file path into string readable by python by using 'r'

df = pd.read_csv(r"C:\Users\revan\Desktop\Book1.csv")
df.head()
```

```python
Offenses=df['Offenses']
total=Offenses.sum()


rate_1=(df.iloc[0,1]/total)*100
rate_2=(df.iloc[1,1]/total)*100
rate_3=(df.iloc[2,1]/total)*100
rate_4=(df.iloc[3,1]/total)*100
rate_5=(df.iloc[4,1]/total)*100
rate_6=(df.iloc[5,1]/total)*100
rate_7=(df.iloc[6,1]/total)*100


labels='White','Black','Other','Unknown','Hispanic','Non-Hispanic','no race given'

slices=[rate_1,rate_2,rate_3,rate_4,rate_5,rate_6,rate_7]

plt.pie(slices, labels=labels, startangle=90, shadow=True, autopct= '% 1.1f %%')
plt.legend()
plt.show()


labels_mf='Male','Female','Unknown'

male=df['Male']
female=df['Female']
unknown=df['Unknown']

utotal=unknown.sum()
mtotal=male.sum()
ftotal=female.sum()

urate=utotal/total*100
frate=ftotal/total*100
mrate=mtotal/total*100

slices_mf=[mrate,frate,urate]

plt.pie(slices_mf, labels=labels_mf, shadow=True, startangle=90, autopct= '% 1.1f %%')
plt.legend()
plt.show()
```
